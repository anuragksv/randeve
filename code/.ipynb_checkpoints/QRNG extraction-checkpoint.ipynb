```python
#package/library imports
import numpy as np
import matplotlib.pyplot as plt
from sympy import fft
from scipy.linalg import toeplitz
```

```python
# Python program to convert a list to string  
def listToString(k): 
    
    # initialize an empty string
    key = "" 
    
    # traverse in the string  
    for element in k: 
        key += element  
    
    # return string  
    return key 
```
